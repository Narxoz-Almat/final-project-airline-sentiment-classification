```python
import numpy as np
```

```python
#A.1 Matrix Multiplication(Resource Allocation)
Resources_Matrix = np.array([[100, 200],
                             [150, 250],
                             [120, 180]])

Allocation_Factors = np.array([[0.8, 1.2],
                               [1.0, 1.5]])

Result_Matrix = Resources_Matrix @ Allocation_Factors
print("A.1 Resulting Matrix:\n", Result_Matrix)
```

```python
# A.2 Element-wise Operations (Production Tracking)
Shift_A_Production = np.array([
    [10, 12, 11, 13, 14, 15, 16],
    [8, 9, 10, 11, 12, 12, 13],
    [14, 13, 15, 14, 13, 16, 15]
])

Shift_B_Production = np.array([
    [9, 10, 11, 10, 12, 14, 13],
    [7, 8, 9, 10, 11, 10, 12],
    [12, 13, 14, 13, 14, 15, 14]
])

Total_Production = Shift_A_Production + Shift_B_Production
print("\nA.2 Total Production:\n", Total_Production)
```

```python
# A.3 Activation Function (Sigmoid)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

test_values = np.array([0, 1, -1, 2])
sigmoid_output = sigmoid(test_values)
print("\nA.3 Sigmoid Output:\n", sigmoid_output)
```

```python
# A.4 Gradient Calculation (Sigmoid derivative)
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

gradient_output = sigmoid_derivative(test_values)
print("\nA.4 Sigmoid Gradient:\n", gradient_output)
```

**Part C**

```python
X = np.array([
    [20, 3, 4],
    [15, 5, 3],
    [30, 2, 2],
    [25, 4, 1],
    [35, 2, 3]
], dtype=float)

Y = np.array([[18], [20], [22], [25], [30]], dtype=float)

X = X / np.max(X, axis=0)
Y = Y / np.max(Y)
```

```python
# Activation functions
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

# Initialize weights
np.random.seed(42)
W1 = np.random.randn(3, 3) * 0.1
b1 = np.zeros((3, 1))
W2 = np.random.randn(1, 3) * 0.1
b2 = np.zeros((1, 1))
```

```python
# Training parameters
lr = 0.1
epochs = 5000
m = X.shape[0]

# Training loop
for epoch in range(epochs):
    Z1 = X @ W1.T + b1.T
    A1 = relu(Z1)
    Z2 = A1 @ W2.T + b2
    A2 = sigmoid(Z2)

    loss = np.mean((A2 - Y) ** 2)

    dA2 = 2 * (A2 - Y) / m
    dZ2 = dA2 * sigmoid_derivative(Z2)
    dW2 = dZ2.T @ A1
    db2 = np.sum(dZ2, axis=0, keepdims=True).T
    dA1 = dZ2 @ W2
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = dZ1.T @ X
    db1 = np.sum(dZ1, axis=0, keepdims=True).T

    W1 -= lr * dW1
    b1 -= lr * db1
    W2 -= lr * dW2
    b2 -= lr * db2

    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
```

```python
# Final results
print("\nFinal Weights and Biases:")
print("W1:", W1)
print("b1:", b1)
print("W2:", W2)
print("b2:", b2)

predictions = sigmoid(relu(X @ W1.T + b1.T) @ W2.T + b2)
print("\nPredictions (normalized):")
print(predictions)

predictions_orig = predictions * np.max(Y)
print("\nPredictions (original scale):")
print(predictions_orig)
```
